Calculating the Air Quality Index
Using this formula
Ip = [IHi - ILo / BPHi - BPLo](Cp - BPLo) + ILo

where,
  Ip = index of Pollutant
  Cp = truncated concentration of pollutant p
  BPHi = concentration breakpoint i.e greater than or equal to Cp
  BPLo = concentration breakpoint i.e less than or equal to Cp
  IHi = AQI value corresponding to BPHi
  ILo = AQI value corresponding to BPLo

PM2.5 CALCULATION

```python
def PM25_Index(x):
  Ip = 0
  Cp = x
  BPHi = 0
  BPLo = 0
  IHi = 0
  ILo = 0
  if (0 <= x <= 30):
    BPHi = 30
    BPLo = 0
    IHi = 50
    ILo = 0
    first = (IHi - ILo) / (BPHi - BPLo)
    second = (Cp - BPLo) + ILo
    Ip = first * second
    return Ip
  elif (31 <= x <= 60):
    BPHi = 60
    BPLo = 31
    IHi = 100
    ILo = 51
    first = (IHi - ILo) / (BPHi - BPLo)
    second = (Cp - BPLo) + ILo
    Ip = first * second
    return Ip
  elif (61 <= x <= 90):
    BPHi = 90
    BPLo = 61
    IHi = 200
    ILo = 101
    first = (IHi - ILo) / (BPHi - BPLo)
    second = (Cp - BPLo) + ILo
    Ip = first * second
    return Ip
  elif (91 <= x <= 120):
    BPHi = 120
    BPLo = 91
    IHi = 300
    ILo = 201
    first = (IHi - ILo) / (BPHi - BPLo)
    second = (Cp - BPLo) + ILo
    Ip = first * second
    return Ip
  elif (121 <= x <= 250):
    BPHi = 250
    BPLo = 121
    IHi = 400
    ILo = 301
    first = (IHi - ILo) / (BPHi - BPLo)
    second = (Cp - BPLo) + ILo
    Ip = first * second
    return Ip
  elif(250 <= x <= 500):
    BPHi = 500
    BPLo = 250
    IHi = 500
    ILo = 401
    first = (IHi - ILo) / (BPHi - BPLo)
    second = (Cp - BPLo) + ILo
    Ip = first * second
    return Ip
  else:
    return Ip
```

CALCULATING THE OVERALL AIR QUALITY INDEX

```python
DATAFRAME.loc[:, 'AQI'] = DATAFRAME.apply(lambda row: max(row['PM2.5 Index'], row['PM10 Index'], row['SO2 Index'], row['NO2 Index'], row['CO Index'], row['O3 Index'], row['NH3 Index']),axis=1)
DATAFRAME = DATAFRAME.applymap(lambda x: round(x, 2))
DATAFRAME
```

```python
DATAFRAME = DATAFRAME.drop(['PM2.5 Index', 'PM10 Index', 'SO2 Index', 'NO2 Index', 'CO Index', 'O3 Index', 'NH3 Index'],axis=1)
DATAFRAME
```

REGRESSION MODELLING

```python
INPUTS = DATAFRAME.drop(['AQI'], axis = 1)
OUTPUT = DATAFRAME['AQI']
```

```python
INPUTS_train, INPUTS_test, OUTPUT_train, OUTPUT_test = train_test_split(INPUTS,OUTPUT,test_size=0.2,random_state=42)
print(INPUTS_train.shape)
print(INPUTS_test.shape)
print(OUTPUT_train.shape)
print(OUTPUT_test.shape)
```

LINEAR REGRESSION

```python
regression = LinearRegression()

regression.fit(INPUTS_train,OUTPUT_train)

PREDICTIONS = regression.predict(INPUTS_test)

mse = mean_squared_error(OUTPUT_test,PREDICTIONS)

mae = mean_absolute_error(OUTPUT_test,PREDICTIONS)

r2s = r2_score(OUTPUT_test,PREDICTIONS)

print("The mean square error is: ", mse)

print("The mean absolute squared error is: ", mae)

print("The root squared error is: ", r2s)
```

```python
data = {
    'True Value': OUTPUT_test,
    'Predicted Value': PREDICTIONS
}

df = pd.DataFrame(data)
df
```

LASSO REGRESSION

```python
lasso = Lasso(alpha = 1.0)

lasso.fit(INPUTS_train,OUTPUT_train)

PREDICTIONS = lasso.predict(INPUTS_test)

mse = mean_squared_error(OUTPUT_test,PREDICTIONS)

mae = mean_absolute_error(OUTPUT_test,PREDICTIONS)

r2s = r2_score(OUTPUT_test,PREDICTIONS)

print("The mean square error is: ", mse)

print("The mean absolute squared error is: ", mae)

print("The root squared error is: ", r2s)
```

```python
tree = DecisionTreeRegressor()

tree.fit(INPUTS_train,OUTPUT_train)

PREDICTIONS = tree.predict(INPUTS_test)

mse = mean_squared_error(OUTPUT_test,PREDICTIONS)

mae = mean_absolute_error(OUTPUT_test,PREDICTIONS)

r2s = r2_score(OUTPUT_test,PREDICTIONS)

print("The mean square error is: ", mse)

print("The mean absolute squared error is: ", mae)

print("The root squared error is: ", r2s)
```
